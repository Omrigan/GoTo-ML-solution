```python
def to_genres(dct):
    out = ""
    for i, c in dct.items():
        if i==-1:
            out+="no genre: %s " % c
        else:
            out+="genre #%s: %s " % (i, c)
    return out
```

```python
print("Task #2 part 1: %s" % to_genres(genres))
```

```python
channels_films = collections.defaultdict(lambda:collections.defaultdict(lambda: 0))
a = 0
for s in schedule.iterrows():
    c = str(s[1]['channel'])
    f = str(s[1]['item_id'])
    channels_films[c][films[f]['genre']]+=films[f]['likes']
    channels_films[c][-2]+=films[f]['likes']
    a+=1
    if a%10000==0:
        print(a)
print('Done')
```

```python
channels = sorted(channels_films.items(), key=lambda c: -c[1][-2])
i = 0
cnt = 0
last = -1
while cnt<10:
    c = channels[i]
    total = c[1][-2]
    del c[1][-2]
    if total!=last:
        print("Channel #%s with id '%s'. Total: %s. Other: %s" %(i+1, c[0], total, to_genres(c[1])))
        last = total
        cnt+=1
    i+=1
```

```python
#Coefitients for each combination {True, False}*{Positive, Negative}
TP = 1000 
FP = -1
TN = -1
FN = 0
def quality(T, p):
    return T*p*TP + (1-T)*p*FP + T*(1-p)*TN + FN*(1-T)*(1-p)

def validate(predict_fun, batch_size):
    s_quality = 0
    u_id = random.choice(users_list)
    print(u_id)
    #print(likes[likes.user_id==u_id])
    for i in range(batch_size):
        f_id = random.choice(films_list)
        p = predict_fun(u_id, f_id)
        T = int(f_id in likes[likes.user_id==u_id].item_id)
        s_quality+=quality(T, p)
    return s_quality/batch_size
    
```
